```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Crear figura 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Definir las posiciones de las capas
capas = [12, 16, 8, 1]  # número de neuronas por capa
colores = ['r', 'g', 'b', 'y']

# Dibujar neuronas
for i, num_neuronas in enumerate(capas):
    x = np.ones(num_neuronas) * i
    y = np.arange(num_neuronas)
    z = np.zeros(num_neuronas)
    ax.scatter(x, y, z, c=colores[i], marker='o', s=100, label=f'Capa {i}')

# Dibujar conexiones (simplificado)
for i in range(len(capas)-1):
    for j in range(capas[i]):
        for k in range(capas[i+1]):
            x = [i, i+1]
            y = [j, k]
            z = [0, 0]
            ax.plot(x, y, z, 'k-', alpha=0.1)

ax.set_xlabel('Capas')
ax.set_ylabel('Neuronas')
ax.set_zlabel('')
ax.legend()
plt.title('Arquitectura de la red neuronal en 3D')
plt.show()
```
